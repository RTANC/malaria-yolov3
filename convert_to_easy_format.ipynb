```python
f = open("data/custom/train.txt","r")
files = list(f)
f.close()
result =""
for file in files:
    img_path = file.rstrip()
    label_path = img_path.replace('/images','/labels').replace('.jpg','.txt')
    print(label_path)
    img = Image.open(img_path)
    img_w, img_h = img.size
#     print("%d, %d" % (img_w,img_h))
    result += "\n" + img_path
    f = open(label_path,"r")
    lines = f.readlines()
    f.close()
    for line in lines:
        class_id, xc, yc, bw, bh = line.split(" ")
        xc = float(xc)
        yc = float(yc)
        bw = float(bw)
        bh = float(bh)
        minx = (xc - (bw/2)) * img_w
        miny = (yc - (bh/2)) * img_h
        bw = bw * img_w
        bh = bh * img_h
        maxx = minx + bw
        maxy = miny + bh
        minx = max(round(minx),0)
        miny = max(round(miny),0)
        maxx = min(round(maxx),img_w)
        maxy = min(round(maxy),img_h)
        bbox = " " + str(minx) + "," + str(miny) + "," + str(maxx) + "," + str(maxy) + "," + class_id
        result += bbox

f = open("2012_train.txt","w")
result = result[1:]
f.write(result)
f.close()
```
