```python
import glob
import random
import os
```

```python
def reset(percent=15):
    return random.randrange(100) < percent
```

```python
training_set = glob.glob("data/custom/images/*-train-*.jpg")
# valid_set = glob.glob("data/custom/images/*-test-*.jpg")
# random.shuffle(training_set)
# random.shuffle(valid_set)
```

```python
print(len(training_set))
# print(len(valid_set))
```

```python
result = ""
for line in training_set:
    print(line)
    result += "\n"+line

file1 = open(os.path.join("data","custom","train.txt"),"w")
file1.write(result[1:])
file1.close()
```

```python
# summary training set
f = open("data/custom/train.txt","r")
lines = list(f)
f.close()
count = [0,0,0,0]
for line in lines:
#     print(line)
    file = line.rstrip().replace("/images","/labels").replace(".jpg",".txt")
    print(file)
    f = open(file,'r')
    bbs = list(f)
    f.close()
    for bb in bbs:
        c,cx,cy,bw,bh = bb.split()
        if c == "0":
            count[0] += 1
        elif c == "1":
            count[1] += 1
        elif c == "2":
            count[2] += 1
        elif c == "3":
            count[3] += 1
```

```python
print("Training Set")
print("Negative %d" % count[0])
print("Positive %d" % count[1])
```

```python
pv_test = glob.glob("data/custom/images/pv-test-*.jpg")
pf_test = glob.glob("data/custom/images/pf-test-*.jpg")
pm_test = glob.glob("data/custom/images/pm-test-*.jpg")
po_test = glob.glob("data/custom/images/po-test-*.jpg")
```

```python
result = ""
for line in pv_test:
    if reset():
        result += "\n"+line

for line in pf_test:
    result += "\n"+line
    
for line in pm_test:
    result += "\n"+line

file1 = open(os.path.join("data","custom","valid.txt"),"w")
file1.write(result[1:])
file1.close()
```
